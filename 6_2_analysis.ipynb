```python
import json

with open("results.json", "r") as f:
    data = json.load(f)

print(data)
```

```python
import matplotlib.pyplot as plt

coh = [[el['cohesion'] for el in gen] for gen in data]
sep = [[el['separation'] for el in gen] for gen in data]
al_str = [[el['alignment_strength'] for el in gen] for gen in data]
fitness = [[el['score'] for el in gen] for gen in data]

# Plot histograms
fig, axes = plt.subplots(len(data), 4, figsize=(15, 10))


for i, (cohesion, separation, alignment, fitness) in enumerate(zip(coh, sep, al_str, fitness)):
    axes[i, 0].hist(cohesion, bins=10, color='skyblue', edgecolor='black')
    axes[i, 0].set_title(f'Generation {i+1} - Cohesion')
    axes[i, 0].set_xlim(-.1, .1)
    axes[i, 0].set_xlabel('Cohesion')
    axes[i, 0].set_ylabel('Frequency')

    axes[i, 1].hist(separation, bins=10, color='salmon', edgecolor='black')
    axes[i, 1].set_title(f'Generation {i+1} - Separation')
    axes[i, 1].set_xlim(-.1, .1)
    axes[i, 1].set_xlabel('Separation')
    axes[i, 1].set_ylabel('Frequency')

    axes[i, 2].hist(alignment, bins=10, color='lightgreen', edgecolor='black')
    axes[i, 2].set_title(f'Generation {i+1} - Alignment Strength')
    axes[i, 2].set_xlim(-.1, .1)
    axes[i, 2].set_xlabel('Alignment Strength')
    axes[i, 2].set_ylabel('Frequency')

    axes[i, 3].hist(fitness, bins=10, color='purple', edgecolor='black')
    axes[i, 3].set_title(f'Generation {i+1} - Fitness')
    axes[i, 3].set_xlim(-1, 1)
    axes[i, 3].set_xlabel('Fitness')
    axes[i, 3].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
