```python
import json

with open("results.json", "r") as f:
    data = json.load(f)

print(data)
```

```python
import matplotlib.pyplot as plt

coh = [[el['cohesion'] for el in gen] for gen in data]
sep = [[el['separation'] for el in gen] for gen in data]
al_str = [[el['alignment_strength'] for el in gen] for gen in data]
fitness = [[el['score'] for el in gen] for gen in data]

# Plot histograms
for i, (cohesion, separation, alignment, fitness) in enumerate(zip(coh, sep, al_str, fitness)):
    fig, axes = plt.subplots(1, 4, figsize=(15, 5))
    axes[0].hist(cohesion, bins=10, color='skyblue', edgecolor='black')
    axes[0].set_title(f'Generation {i+1} - Cohesion')
    axes[0].set_xlim(0, 1)
    axes[0].set_xlabel('Cohesion')
    axes[0].set_ylabel('Frequency')

    axes[1].hist(separation, bins=10, color='salmon', edgecolor='black')
    axes[1].set_title(f'Generation {i+1} - Separation')
    axes[1].set_xlim(0, 1)
    axes[1].set_xlabel('Separation')
    axes[1].set_ylabel('Frequency')

    axes[2].hist(alignment, bins=10, color='lightgreen', edgecolor='black')
    axes[2].set_title(f'Generation {i+1} - Alignment Strength')
    axes[2].set_xlim(0, 1)
    axes[2].set_xlabel('Alignment Strength')
    axes[2].set_ylabel('Frequency')

    axes[3].hist(fitness, bins=10, color='purple', edgecolor='black')
    axes[3].set_title(f'Generation {i+1} - Fitness')
    axes[3].set_xlim(0.5, 1)
    axes[3].set_xlabel('Fitness')
    axes[3].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
